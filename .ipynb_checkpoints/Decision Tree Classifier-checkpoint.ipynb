## Importing the libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Importing the dataset:

```python
data= pd.read_csv('bank.csv')
data.head()
```

## Domain analysis:

- The target variable is 'deposit' indicating whether the client subscribed to a term deposit or not.
- Input variables: age, job, marital, education, default( has credit in default), 
- balance- average yearly balance, in euros 
- housing- has housing loan or not
- loan- whether the client has a personal loan.
 #### related with the last contact of the current campaign:
- contact- communication type used to contact the client
- day- last contact day of the month
- month- last contact month of the year
- duration- duration of the last contact in seconds
#### other attributes:
- campaign- Number of contacts performed during this campaign for the client.
- pdays-  Number of days that passed by after the client was last contacted from a previous campaign.
- previous- Number of contacts performed before this campaign for the client.
- poutcome- Outcome of the previous marketing campaign.
 
- The data provides information about the telemarketing campaign undertaken by the bank, information about the clients such as their behavioural data and economic indicators.

## Exploratory Data Analysis: 

### Univariate analysis:

```python
# numerical features-  age, balance, day, duration, campaign, pdays, previous  
# categorical features- job, marital, education, default, housing, loan, contact, month, poutcome, deposit
```

```python
data_num = data[['age', 'balance', 'day', 'duration','campaign','pdays', 'previous']]
data_cat= data[['job','marital','education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'deposit']]
```

```python
# Numerical columns- totally 7

plt.figure(figsize=(15,20), facecolor = 'white')
plotnumber =1

for column in data_num:
    if plotnumber <=8:
        ax= plt.subplot(4,2, plotnumber)
        sns.histplot(data=data_num, x= column, kde = True)
        plt.xlabel(column, fontsize= 15)
        plt.ylabel('count', fontsize =15)
    plotnumber +=1
plt.tight_layout()
```

#### Insights:
- Most of the clients are from the age group of 25-45 range and the distribution is skewed.
- Almost 75% fall in the balance range of negative balance to 10000 and the distributi0n is highly skewed.
- 'day' variable shows uneven distribution. It is skewed.
- Almost 70% of records occupy the contact duration of range of 0-500seconds and the distribution is skewed.
- About 90% of clients been contacted for 0-10 times during this campaign.
- It has been around 0-50 number of days passed by after the client was last contacted from a previous campaign for almost 70% of clients.
- Less than one contact has been made for almost 80% of the clients.

```python
# Categorical columns- totally 10

plt.figure(figsize=(30,25), facecolor = 'white')
plotnumber =1

for column in data_cat:
    if plotnumber <=10:
        ax= plt.subplot(5,2, plotnumber)
        sns.countplot(data=data_cat, x= column)
        plt.xlabel(column, fontsize= 20)
        plt.ylabel('count', fontsize =20)
    plotnumber +=1
plt.tight_layout()
```

#### Insights:
- Management occupies the highest position followed by the blue collar job in the job type.
- Most of the clients are married.
- In Education level, secondary level is the dominant position.
- Amost 90% has not defaulted in credit.
- Cellular contact is performed for most of the clients.
- May month is the most contacted month followed by July and August.
- The outcome of the previous campaign is unknown for most of the records.
- As far as the target variable 'deposit', almost 60% of clients have not subscribed to 
the term deposit.

### Bivariate analysis: 

#### Relationship between numerical columns and the target variable. 

```python
# Numerical columns- totally 7, target variable-deposit (categorical)

plt.figure(figsize=(15,20), facecolor = 'white')
plotnumber =1

for column in data_num:
    if plotnumber <=8:
        ax= plt.subplot(4,2, plotnumber)
        sns.barplot(data=data_num, x=  data['deposit'], y= column, hue= data['deposit'])
        plt.xlabel('deposit', fontsize= 15)
        plt.ylabel(column, fontsize =15)
    plotnumber +=1
plt.tight_layout()
```

#### Insights:
- Despite age differences, clients have equal probability of subscribing and not subscribing to the term deposits.
- Clients who have more than 1250 euros as balance in their accounts are more likely to subscribe to the term deposits and those who have less than that are more probably not to subscribe to the term deposits.
- If the duration of the contact is more than 200 seconds, then those clients are more likely to subscribe to the term deposits.
- More the number of contacts performed during the campaign, less the subscribing rate of clients.
- More the no of days passed by after the last contact was made during the previous campaign, the clients are more likely to subscribe to the term deposits.
- If more number of contacts are made during the previous campaign, the clients are more likely to subscribe to the term deposits.

### Encoding: 

```python
data_cat.columns
```

```python
# 1. job
data['job'].value_counts() # from management to unknown- decreasing order
```

```python
data['job'] = data['job'].map({'unknown':0, 'housemaid':1,
                               'entrepreneur':2, 'unemployed':3, 'student':4, 'self-employed':5,
                               'retired':6, 'services':7,'admin.':8, 'technician':9,'blue-collar':10,
                               'management':11})
```

```python
data[['job']].head(5)
```

```python
# 2. marital
data['marital'].value_counts() 
```

```python
data['marital'] = data['marital'].map({'divorced':0,  'single':1, 'married':2}) #from married to divorced- decreasing order
```

```python
data[['marital']].head()
```

```python
#3. education
data['education'].value_counts() 
```

```python
data['education'] = data['education'].map({'unknown':0, 'primary':1, 'secondary':2, 'tertiary':3})
```

```python
data[['education']].head()
```

```python
#4. default
data['default'].value_counts()
```

```python
data['default'] = data['default'].map({'yes':1, 'no':0})
```

```python
data[['default']] .head()
```

```python
# 5. housing
data['housing'].value_counts()
```

```python
data['housing'] = data['housing'].map({'yes':1, 'no':0})
```

```python
data[['housing']].head()
```

```python
# 6. loan
data['loan'].value_counts()
```

```python
data['loan'] = data['loan'].map({'yes':1, 'no':0})
```

```python
data[['loan']].head()
```

```python
# 7.contact
data['contact'].value_counts()
```

```python
data['contact'] = data['contact'].map({'cellular':2, 'telephone':1, 'unknown':0})
```

```python
data[['contact']]
```

```python
# 8. month
data['month'].value_counts()
```

```python
data['month'] =data['month'].map({'dec':11, 'nov':10, 'oct':9, 'sep':8, 'aug':7, 'jul':6, 'jun':5, 'may':4, 'apr':3, 'mar':2, 'feb':1, 'jan':0})
```

```python
data['month'].head()
```

```python
#9. poutcome
data.poutcome.value_counts()
```

```python
data['poutcome'] = data['poutcome'].map({'success':3, 'failure':2, 'other':1, 'unknown':0})
```

```python
data.poutcome.head()
```

```python
#10. deposit
data.deposit.value_counts()
```

```python
data.deposit= data.deposit.map({'yes':1, 'no':0})
```

```python
data.deposit.head()
```

#### Balancing the dataset: 

```python
# splitting the data
X= data.drop('deposit', axis =1)
y = data['deposit']
```

```python
y.value_counts() # imbalanced target variable
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state =14)
```

```python
# using SMOTE function
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

```python
from collections import Counter

print(Counter(y_train))
print(Counter(y_train_sm))
```

##  Model Creation:

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()

model_dt.fit(x_train_sm, y_train_sm)
```

```python
# prediction
y_predict= model_dt.predict(x_test)
y_predict
```

```python
# evaluation

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# training accuracy
y_train_acc = model_dt.predict(x_train_sm)
accuracy_score(y_train_sm, y_train_acc)
```

## Hyperparameter tuning:

```python
from sklearn.model_selection import GridSearchCV

model = DecisionTreeClassifier()

params = {'criterion': ('gini', 'entropy'),
         'max_depth': list(range(1,20)),
         'min_samples_split': [2, 3, 4], 
         'min_samples_leaf': list(range(1, 20))}
model_for_cv = GridSearchCV(model, params, scoring='f1', verbose=1, n_jobs=-1, cv= 3)

model_for_cv.fit(x_train_sm, y_train_sm)

best_params_ = model_for_cv.best_params_
print(f'Best paramaters: {best_params_}')
```

#### Applying the best hyperparameters to the model and training : 

```python
model1 = DecisionTreeClassifier(criterion= 'gini', max_depth=  11,
                                min_samples_leaf= 13, min_samples_split= 4)

model1.fit(x_train_sm, y_train_sm)
```

```python
#prediction

y_predict1= model1.predict(x_test)
y_predict1 
```

```python
#evaluation

accuracy_score(y_test, y_predict1)
```

```python
confusion_matrix(y_test, y_predict1)
```

```python
print(classification_report(y_test, y_predict1))
```
